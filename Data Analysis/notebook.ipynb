```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("Data/06", "r")

N = int(file.readline())
K = int(file.readline())
T = int(file.readline())
R = int(file.readline())

s0 = [[[[0 for t in range(T)] for n in range(N)] for r in range(R)] for k in range(K)]
d = [[[[0 for n in range(N)] for m in range(R)] for r in range(N)] for k in range(K)]


for t in range(T):
    for k in range(K):
        for r in range(R):
            values = file.readline().strip().split()
            for n in range(N):
                s0[k][r][n][t] = float(values[n])


for k in range(K):
    for r in range(R):
        for m in range(N):
            values = file.readline().strip().split()
            for n in range(N):
                d[k][m][r][n] = float(values[n])
```

```python
for k in range(K):
    for r in range(R):
        for n in range(N):
            val = []
            for t in range(T):
                val.append(s0[k][r][n][t])
            
            # Usamos plt.plot para graficar val a lo largo del tiempo
            plt.plot(range(T), val)  # range(T) crea una lista de tiempos [0, 1, 2, ..., T-1]
            
            plt.xlabel('Tiempo')  # Etiqueta para el eje x
            plt.ylabel('Valor')  # Etiqueta para el eje y
            plt.title(f'Serie temporal para k={k}, r={r}, n={n}')  # Título del gráfico
            plt.grid(True)  # Añade una cuadrícula al gráfico para facilitar la lectura
            plt.show()  # Muestra el gráfico
            

# for k in range(K):
#     for r in range(R):
#         val = []    
#         for m in range(N):
#             for n in range(N):
#                 val.append(d[k][m][r][n])

#         plt.hist(val, bins=20, edgecolor='black', range=(-2, 0))

#         plt.show()
```
